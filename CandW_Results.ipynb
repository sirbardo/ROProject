```python
import os
from tabulate import tabulate
```

```python
def get_table(path):
    table = []
    files = os.listdir(path)
    files.sort()
    for filename in files:
        with open(path+filename,'r') as f:
            content = f.read()
        content = content.split('\n')
        of = float(content[0].split(' ')[2])
        time = float(content[1].split(' ')[3])+float(content[4].split(' ')[4])

        with open('Instances/'+filename,'r') as f:
            content = f.read()
        content = content.split('\n')
        n_of_customers = int(content[0])

        with open('Instances/'+filename,'r') as f:
            content = f.read()
        content = content.split('\n')
        gap = of - float(content[9].split(' ')[3])

        table.append([filename[0:2],n_of_customers,of,time,gap])
    
    return table
```

```python
table1 = get_table('parallel_routes/')
print(tabulate(table1,headers=['instance name','customers','objective value','time','cost gap'],tablefmt='grid',stralign='center',numalign='center'))
```

```python
table2 = get_table('sequential_routes/')
print(tabulate(table2,headers=['instance name','customers','objective value','time','cost gap'],tablefmt='grid',stralign='center',numalign='center'))
```

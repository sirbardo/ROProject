```python
import os
from tabulate import tabulate
```

```python
def get_table(path):
    table = []
    files = os.listdir(path)
    files.sort()
    for filename in files:
        
        if '.txt' not in filename:
            continue
        
        with open(path+filename,'r') as f:
            content = f.read()
        content = content.split('\n')
        of = float(content[0].split(' ')[2])
        time = float(content[1].split(' ')[3])+float(content[2].split(' ')[4])

        with open('Instances/'+filename,'r') as f:
            content = f.read()
        content = content.split('\n')
        n_of_customers = int(content[0])

        with open('DetailedSols/RPA_Solutions/Detailed_Solution_'+filename,'r') as f:
            content = f.read()
        content = content.split('\n')
        opt_val = float(content[8].split(' ')[3])
        gap = of - opt_val

        table.append([filename[0:2],n_of_customers,of,time,gap])
    
    return table
```

```python
table1 = get_table('parallel_routes/')
print(tabulate(table1,headers=['instance name','customers','objective value','time','cost gap'],tablefmt='grid',stralign='right',numalign='decimal'))
with open('parallel_resuts.txt','w') as f:
    f.write(tabulate(table1,headers=['instance name','customers','objective value','time','cost gap'],tablefmt='grid',stralign='right',numalign='decimal'))
```

```python
table2 = get_table('sequential_routes/')
print(tabulate(table2,headers=['instance name','customers','objective value','time','cost gap'],tablefmt='grid',stralign='right',numalign='decimal'))
with open('sequential_resuts.txt','w') as f:
    f.write(tabulate(table2,headers=['instance name','customers','objective value','time','cost gap'],tablefmt='grid',stralign='right',numalign='decimal'))
```
